```python
import ipywidgets as widgets
```

```python
ngram = widgets.Dropdown(
    options=['Unigrams', 'Bigrams'],
    value="Unigrams",
    description='N-Gram:',
    disabled=False,
)
size = widgets.Dropdown(
    options=[30,100, 500],
    value=100,
    description='Size:',
    disabled=False,
)
start_date = widgets.Dropdown(
    options=['01-01-2010', '01-01-2019'],
    value='01-01-2010',
    description='Start Date:',
    disabled=False,
)
end_date = widgets.Dropdown(
    options=['30-06-2019'],
    value='30-06-2019',
    description='End Date:',
    disabled=False,
)
num_topics = widgets.Dropdown(
    options=[6, 12, 24],
    value=12,
    description='Num Topics:',
    disabled=False,
)
```

```python
display(ngram)
display(size)
display(start_date)
display(end_date)
display(num_topics)
```

```python
column_name = ngram.value
gram = {'Unigrams':1, 'Bigrams': 2}[column_name]
filename = f'./models/lda_{column_name}_{size.value}_d{start_date.value}d_d{end_date.value}d_{num_topics.value}'
filename
```

```python
from gensim.models import LdaModel, LsiModel, HdpModel
import pickle
```

```python
with open(filename+'.model', 'rb') as handle:
    lda_model = pickle.load(handle)
with open(filename+'.corpus', 'rb') as handle:
    corpus = pickle.load(handle)
with open(filename+'.dict', 'rb') as handle:
    dictionary = pickle.load(handle)
```

```python
import pyLDAvis.gensim
import time
```

```python
pyLDAvis.enable_notebook()
```

```python
start = time.time()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
end = time.time()
print(f'Visualization generation finished in {end-start:.2f} seconds.\n')
```
